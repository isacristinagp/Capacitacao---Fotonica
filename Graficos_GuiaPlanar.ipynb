```python
import numpy as np
import matplotlib.pyplot as plt

nc = 1
nf = 3.5
ns = 1.45

a = 0.25e-6
lam = 1550e-9
k0 = 2*np.pi/lam

neff_vals = np.linspace(nc, nf, 600)

u = a*k0*np.sqrt(nf**2 - neff_vals**2)
v = a*k0*np.sqrt(neff_vals**2 - ns**2)
w = a*k0*np.sqrt(neff_vals**2 - nc**2)

LHS = u*(v + w) / (u**2 - v*w)
LHS[np.abs(LHS) > 25] = np.nan
RHS = np.tan(2*u)
RHS[np.abs(RHS) > 20] = np.nan

plt.ylim(-4, 4)
plt.xlim(1.5, 3.5)
plt.plot(neff_vals, LHS, label="LHS")
plt.plot(neff_vals, RHS, label="RHS")
plt.xlabel("n_eff")
plt.ylabel("LHS e RHS")
plt.grid(True)
plt.legend()
plt.show()
```

```python
delta = (ns**2 - nc**2)/(nf**2 - ns**2)
m = [0, 1, 2, 3, 4]

b = np.linspace(0, 1, 100)

V = lambda m: (m*np.pi + np.arctan (np.sqrt((b+delta)/(1-b))) + np.arctan (np.sqrt((b)/(1-b))))/(2*np.sqrt(1-b))

for i in m:
    plt.plot(V(i), b)

plt.xlim(0,6)
plt.xlabel('V - frequência normalizada')
plt.ylabel('b - constante de propagação normalizada')
plt.show()
```
