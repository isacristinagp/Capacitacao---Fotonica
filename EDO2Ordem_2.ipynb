```python
import numpy as np
import matplotlib.pyplot as plt

def solucao_analitica(t):
  A = 1
  B = 0
  return A*np.cos(2*t)+B*np.sin(2*t)


h = 0.1
w=2
ti = 0
tf = 10
y0 = 1
y1 = y0
y = [y0, y1]
t = [ti, ti+h]
y_an = [solucao_analitica(t[0]), solucao_analitica(t[1])]

num_step = int((tf-ti)/h)
for i in np.arange(1, num_step):
  y_proximo = 2*y[i]*(1-w**2*h**2/4)/(1+w**2*h**2/4) - y[i-1]
  t_proximo = t[i]+h
  y_an_prox = solucao_analitica(t_proximo)

  y.append(y_proximo)
  t.append(t_proximo)
  y_an.append(y_an_prox)

plt.figure(figsize=(8, 5))
plt.plot(t, y, label = 'Solução numérica', linewidth=3)
plt.plot(t, y_an, label='Solução analítica', linewidth=3, linestyle='--')
plt.legend()
plt.title('Solução numérica para $y_{n+1}$')
plt.show

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def solucao_analitica(t):
  A = 1
  B = 0
  return A*np.cos(2*t)+B*np.sin(2*t)


def solucao_numerica(h):
  w=2
  ti = 0
  tf = 10
  y0 = 1
  y1 = y0
  y = [y0, y1]
  t = [ti, ti+h]
  y_an = [solucao_analitica(t[0]), solucao_analitica(t[1])]
  erro = [(y_an[0]-y[0])**2, (y_an[1]-y[1])**2]

  num_step = int((tf-ti)/h)
  for i in np.arange(1, num_step):
    t_proximo = t[i]+h
    y_proximo = 2*y[i]*(1-w**2*h**2/4)/(1+w**2*h**2/4) - y[i-1]
    y_an_prox = solucao_analitica(t_proximo)
    erro.append((y_an_prox-y_proximo)**2)


    y.append(y_proximo)
    t.append(t_proximo)
    y_an.append(y_an_prox)

  plt.figure(figsize=(8, 5))
  plt.plot(t, y, label = 'Solução numérica', linewidth=3)
  plt.plot(t, y_an, label='Solução analítica', linewidth=3, linestyle='--')
  plt.legend()
  plt.title('Solução numérica para $y_{n+1}$')
  plt.show

  erro_total = 0
  for i in erro:
    erro_total+=i

  e = math.sqrt(erro_total)/num_step

  return e
h1 = 0.1
h2 = 0.4
e1 = solucao_numerica(h1)
e2 = solucao_numerica(h2)

r = np.log(e2/e1)/np.log(h2/h1)

print(r)
```
