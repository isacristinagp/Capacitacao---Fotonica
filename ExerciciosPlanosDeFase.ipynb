```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


# Funo da soluo analtica
def y_analitica(t):
    return (7/3)*np.exp(-t) - (4/3)*np.exp(-4*t)

def y_derivada(t):
    return (-7/3)*np.exp(-t) + (16/3)*np.exp(-4*t)

# Intervalo de tempo
t = np.linspace(0, 10, 500)
y = y_analitica(t)
y_prime_vals = y_derivada(t)


# Plot
plt.figure(figsize=(8, 5))
plt.plot(t, y, label=r'$y(t) = \frac{7}{3}e^{-t} - \frac{4}{3}e^{-4t}$', color='purple')
plt.title("Soluo Analtica da EDO: y'' + 5y' + 4y = 0")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.legend()
plt.show()

# Sistema convertido: y1 = y, y2 = y'
def sistema(t, v):
    y1, y2 = v
    dy1 = y2
    dy2 = -5*y2 - 4*y1
    return [dy1, dy2]

# Gerao de malha
y1_vals = np.linspace(-3, 3, 25)
y2_vals = np.linspace(-3, 3, 25)
Y1, Y2 = np.meshgrid(y1_vals, y2_vals)

# Campo vetorial
U = Y2
V = -5*Y2 - 4*Y1

# Normaliza os vetores para melhor visualizao
N = np.sqrt(U**2 + V**2)
U_norm = U / N
V_norm = V / N

# Plot
plt.figure(figsize=(8, 6))
plt.quiver(Y2, Y1, V_norm , U_norm, color='blue', alpha=0.7)
plt.plot(y_prime_vals, y, color='blue', label="trajetria analtica")
plt.title("Plano de Fase da EDO: y'' + 5y' + 4y = 0")
plt.xlabel("v")
plt.ylabel("y")
plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```
